```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('historical_weather.csv')
df.head()
```

```python
df.isna().sum()
```

```python
# Fill missing values with mean
df['avg_wind_speed_kmh'].fillna(df['avg_wind_speed_kmh'].mean(), inplace=True)

# guessing the missing values are forwarded
df['min_temp_c'].fillna(method='ffill', inplace=True)
df['max_temp_c'].fillna(method='ffill', inplace=True)
df['avg_temp_c'].fillna(method='ffill', inplace=True)
df['avg_wind_dir_deg'].fillna(method='ffill', inplace=True)
df['min_temp_c'].fillna(method='ffill', inplace=True)
df['max_temp_c'].fillna(method='ffill', inplace=True)

# guessing the missing values are 0
df['precipitation_mm'].fillna(0, inplace=True)
df['snow_depth_mm'].fillna(0, inplace=True)

df.head(10)
```

```python
df_shrunk = df[['city_id', 'date', 'avg_temp_c']]
df_shrunk.head(10)
```

```python
df_shrunk['date'] = pd.to_datetime(df_shrunk['date'])
df['date'] = pd.to_datetime(df['date'])

df_shrunk.head()
```

```python
df_shrunk.set_index('date', inplace=True)
df_shrunk.sort_values(by=['city_id', 'date'], inplace=True)

df_shrunk.head()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import xgboost as xgb
```

```python
def xgboost_model_function(df, city_id):
    city_data = df[df['city_id'] == city_id].copy()

    city_data['date_int'] = city_data.index.astype('int64') // 10**9 
    X = city_data['date_int'].values.reshape(-1, 1)
    y = city_data['avg_temp_c'].values

    train_size = int(len(city_data) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    xgboost_model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)
    xgboost_model.fit(X_train, y_train)

    city_data['xgboost_pred'] = xgboost_model.predict(X)

    mse = mean_squared_error(city_data['avg_temp_c'], city_data['xgboost_pred'])
    print(f'Mean Squared Error for {city_id}: {mse}')

    return xgboost_model, city_data, mse
```

```python
unique_cities = df_shrunk['city_id'].unique()
city_models = {}
predictions = []

for city_id in unique_cities:
    xgboost_model, city_data, mse = xgboost_model_function(df_shrunk, city_id)
    city_models[city_id] = xgboost_model
```

```python
df_key = pd.read_csv('submission_key.csv')
df_key.head(10)
```

```python
df_key['date'] = pd.to_datetime(df_key['date'])
```

```python
for index, row in df_key.iterrows():
    city_id = row['city_id']
    date_int = int(row['date'].timestamp()) 
 
    xgboost_model = city_models[city_id]
    
    xgboost_forecast = xgboost_model.predict(np.array([[date_int]]))[0]

    predictions.append({
        'submission_ID': row['submission_ID'],
        'avg_temp_c': xgboost_forecast
    })

predictions_df = pd.DataFrame(predictions)
predictions_df.head(10)
```
